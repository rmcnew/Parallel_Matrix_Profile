```
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import datetime as dt
```

```
AAPL_df = pd.read_csv("AAPL.csv")
AAPL_df['Date'] = pd.to_datetime(AAPL_df['Date'])

plt.figure(figsize=(14, 5))
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.suptitle('Apple Stock Price', fontsize='30')
plt.xlabel('Date', fontsize ='20')
plt.ylabel('Adj. Close Price', fontsize='20')
plt.plot(AAPL_df['Date'], AAPL_df['Adj Close'].values)
plt.show()
```

```
AAPL_mp_df = pd.read_csv("AAPL-matrix_profile.csv", header=None)
plt.figure(figsize=(14, 5))
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.suptitle('Apple Stock Price Matrix Profile', fontsize='30')
plt.xlabel('Index', fontsize ='16')
plt.ylabel('z-normalized Euclidean distance', fontsize='16')
plt.plot(AAPL_mp_df[0].values)
plt.show()
```
